```python
Hospitals = 25 
Schools = 66
Dist = 231
html =f'''
<table>\
<tr>\
<h5 style="text-align:center;"> \

<b><font color="blue">Dist. &nbsp{Dist} </font</td>\
</h4>\
</td>\

<tr>\
<td><B>Hospitals</td>\
<td>&nbsp&nbsp{Hospitals}</td>\
<td>&nbsp&nbsp&nbsp&nbsp</td>\
<td><img ALIGN="Right" width=37, height=37 src="data:image/png;base64,{Hosp_img}">
</td>\

<tr class="blank_row"> \
    <td colspan="3">&nbsp\
    </td>\
</tr>\

</tr>\
<tr>\
<td><B>Schools</td>\
<td>{Schools}</td>\
<td>&nbsp&nbsp&nbsp&nbsp</td>\
<td><img ALIGN="Right" width=37, height=37 src="data:image/png;base64,{Schol_img}">
</tr>\
</table>
'''
```

```python
#html 
# Read this for the icons
# https://stackoverflow.com/questions/53721079/python-folium-icon-list
# https://stackoverflow.com/questions/58607693/how-to-use-folium-icon-with-fontawesome

# Also read this
help(folium.Icon)
```

```python
District = 234
Coordinates = '40.693, -73.985'
my_test = folium.Map([40.693, -73.985], zoom_start=12)

icon = folium.Icon(color="blue", icon="cloud", icon_color='white')



iframe = folium.IFrame(html, width=200, height=140)
popup = folium.Popup(iframe, max_width=650)
folium.Marker([40.693, -73.985], popup=Coordinates, tooltip=html, icon=icon).add_to(my_test)

# Use prefix='fa' for font-awesome
# https://fontawesome.com/v4.7.0/icons/
# https://fontawesome.com/icons?d=gallery


folium.Marker([40.72, -73.99], icon=folium.Icon(color='red', icon='fa-train', prefix='fa')).add_to(my_test)


folium.Marker([40.72, -73.85], icon=folium.Icon(color='black', icon='fa-graduation-cap', prefix='fa')).add_to(my_test)

folium.Marker([40.69, -73.85], icon=folium.Icon(color='orange', icon='fa-ambulance', prefix='fa')).add_to(my_test)


# Use 'glyphicon' for bootstrap 3.
# https://getbootstrap.com/docs/3.3/components/

folium.Marker([40.65, -73.99], icon=folium.Icon(color='darkgreen', glyphicon='glyphicon glyphicon-asterisk')).add_to(my_test)
```
